```python
from threading import Thread
def suma():
    return 2
def worker(msg):
    """
    Esta función se ejecuta en otro hilo.
    """
    for i in range(10000):
        print(msg,suma())

t = Thread(target=worker, args=("Hola mundo!",))
t1 = Thread(target=worker, args=("Hola david!",))
t.start()
t1.start()
```
